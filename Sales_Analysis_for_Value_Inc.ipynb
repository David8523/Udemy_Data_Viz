## 2. Data Exploration and Manipulation

### 2.1 Importing the libraries and visualizing the first 5 rows


```python
#importing the required librarie
import pandas as pd
```

```python
#loading dataset into dataframe
#from google.colab import files
#uploaded = files.upload()
data = pd.read_csv("transaction.csv")
data.head(5)
```

```python
#making a table with separation between cells
data = pd.read_csv("transaction.csv",sep=";")
data.head(5)
```

### 2.2 Creating new relevant columns
Creation of 4 new numerical columns in order to expand the fiancial information per transaction:
* Cost per transaction: Obtained by multiplying the cost of the product by the number of products sold.
* Sales per transaction: Sales price of the product multiplied by the quantity sold.
* Profit per transaction: Subtraction of sales and costs per transaction.
* Markup: Sales minus costs divided by costs.

```python
#CostPerTransaction Column Calculation
CostPerItem = data ['CostPerItem']
NumberOfItemsPurchased = data ['NumberOfItemsPurchased']
#Cost per transaction
data['CostPerTransaction'] = CostPerItem * NumberOfItemsPurchased
#Sales per Transaction
data['SalesPerTransaction'] = data['SellingPricePerItem'] * data['NumberOfItemsPurchased']
#Profit calculation = sales - cost
data['ProfitPerTransaction'] = data['SalesPerTransaction'] - data['CostPerTransaction']
#Markup = (sales-cost)/Cost
data['Markup'] = (data['SalesPerTransaction'] - data['CostPerTransaction']) / data['CostPerTransaction']
```

### 2.3 Rounding Markup

```python
#rounding Markup
data['Markup'] = round(data['Markup'], 3)
```

### 2.4 Combining dates in one column

Combining the columns Day, Month and Year in one column


```python
#changing data type (month not required)
day = data['Day'].astype(str)
year = data['Year'].astype(str)
data['date'] = day+'-'+data['Month']+'-'+ year
data.head(5)
# so far, so good
```

### 2.5 Splitting Client Keywords in various more detailed columns
We can divide the keywords in 3 more detailled new columns
1. Client Age
2. Client type
3. Type of contract

```python
#using spit to plit the client keywoards fiels
split_col = data['ClientKeywords'].str.split(',', expand=True)
#creating new columns for the split columns in client Keywords
data['ClientAge']=split_col[0]
data['ClientType']=split_col[1]
data['ClientContract']=split_col[2]

#using the replace function to clean the new columns
data['ClientAge']=data['ClientAge'].str.replace('[' , '')
data['ClientContract']=data['ClientContract'].str.replace(']' , '')
```

### 2.6 Merge files with another CSV file
Merge with another crucial file that displays the value of the seasons, creating a correlation between season and sales volume

```python

#from google.colab import files
#uploaded = files.upload()
seasons=pd.read_csv('value_inc_seasons.csv', sep=';')
seasons.head(5)
```

```python
#mergin the files, creating a new dataset
data=pd.merge(data, seasons, on = 'Month')
```

### 2.7 Deleting irrelevant columns
columns like day, month, year or the keywords

```python
#dropping columns
data=data.drop('ClientKeywords', axis=1)
data=data.drop(['Day','Month','Year'], axis=1)
```


### 2.8 Lowercase the products description

Last but no least, lets change the description of the products to lowercase.

```python
#using the lower function to change item to lowercase
data['ItemDescription']=data['ItemDescription'].str.lower()
data.head(5)
```
